```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('image.png')
```

```python
plt.imshow(img)
```

```python
imgrgb=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(imgrgb)
```

```python
imggray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(imggray,cmap='gray')
```

## blob detection using cv2s simpleBlobDetector

```python
params = cv2.SimpleBlobDetector_Params()

params.filterByArea = True
params.minArea = 100
params.filterByCircularity = False
params.filterByConvexity = False
params.filterByInertia = False

detector = cv2.SimpleBlobDetector_create(params)
```

```python
keypoints = detector.detect(img)
```

```python
img_with_keypoints = cv2.drawKeypoints(imggray, keypoints, np.array([]), (255, 0, 0), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
```

```python
plt.imshow(img_with_keypoints)
```

## For colored images

simple Blob detector allows only simgle channels thus we use grayscale 

one way is to use grayscale to get keypoints and then show on colored image

or use individual channels 

```python
image = cv2.imread('image.png')

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

params = cv2.SimpleBlobDetector_Params()

params.filterByArea = True
params.minArea = 100
params.filterByCircularity = False
params.filterByConvexity = False
params.filterByInertia = False

detector = cv2.SimpleBlobDetector_create(params)
keypoints = detector.detect(gray_image)
output_image = cv2.drawKeypoints(image, keypoints, None, (255, 0, 0), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

plt.figure(figsize=(10, 5))
plt.imshow(cv2.cvtColor(output_image, cv2.COLOR_BGR2RGB))  
plt.title("Color Image with Detected Blobs")
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt


image = cv2.imread('image.png')
blue_channel, green_channel, red_channel = cv2.split(image)


params = cv2.SimpleBlobDetector_Params()

params.filterByArea = True
params.minArea = 100
params.filterByCircularity = False
params.filterByConvexity = False
params.filterByInertia = False

detector = cv2.SimpleBlobDetector_create(params)


keypoints_blue = detector.detect(blue_channel)
keypoints_green = detector.detect(green_channel)
keypoints_red = detector.detect(red_channel)

output_image = image.copy()
output_image = cv2.drawKeypoints(output_image, keypoints_blue, None, (255, 0, 0), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)  
output_image = cv2.drawKeypoints(output_image, keypoints_green, None, (0, 255, 0), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)  
output_image = cv2.drawKeypoints(output_image, keypoints_red, None, (0, 0, 255), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)  

plt.figure(figsize=(10, 5))
plt.imshow(cv2.cvtColor(output_image, cv2.COLOR_BGR2RGB))  
plt.title("Color Image with Detected Blobs")
plt.axis('off')
plt.show()
```
